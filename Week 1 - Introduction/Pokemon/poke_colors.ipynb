```python
import urllib3
from bs4 import BeautifulSoup

http = urllib3.PoolManager()
r = http.request('GET', 'bulbapedia.bulbagarden.net/wiki/List_of_Pokemon_by_National_Pokedex_number')
page = r.data

page
```

```python
soup = BeautifulSoup(page, "lxml")

urls = {}
for td in soup.find_all('td'):
    if '_(Pok%C3%A9mon)' in str(td):
        hrf = td.a['href']
        if hrf not in urls:
            name = hrf.split('/')[2].split('_')[0]
            urls[name] = hrf
urls
```

```python
import csv

reader = csv.reader(open('pokes.txt'))
base = 'bulbapedia.bulbagarden.net'

for row in reader:
    r = http.request('GET', base + urls.get(row[1]))
    print(r.data)
    break
```

```python
http = urllib3.PoolManager()
r = http.request('GET', 'http://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_color')
col_page = r.data

col_page
```

```python
col_soup = BeautifulSoup(col_page, "lxml")

cols = [c.text for c in col_soup.find_all('span', {'class': 'mw-headline'})]
print(','.join(cols))

col_diag = [[td.text.replace('\n', '').replace(' ', '') for td in table.find_all('td') if td.text.replace('\n', '').replace(' ', '')] for table in col_soup.find_all('table')]
col_diag = col_diag[:-3]
```

```python
col_d = {poke: cols[i] for i, row in enumerate(col_diag) for poke in row}
col_d
```

```python
reader = csv.reader(open('pokes.txt'))
writer = csv.writer(open('pokec.txt', 'w'))

for row in reader:
    if col_d.get(row[1]):
        row.append(col_d[row[1]])
        writer.writerow(row)
```
