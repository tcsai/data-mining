```python
import csv
reader = csv.reader(open('train.csv'))

for i, row in enumerate(reader):
    if i:
        row[3] = row[3].replace(',', ' -')
        if not row[3].startswith('"'):
            row[3] = '"' + row[3].replace('"', "'") + '"'
        row[3] = row[3].replace(',', ' -')
        row = ['"' + cell + '"' if ' ' in cell and not '"' in cell else cell for cell in row]
        out = ','.join(row).replace(',,', ',?,')
        if out.endswith(','):
            out = out + '?'
        print(out)
```
